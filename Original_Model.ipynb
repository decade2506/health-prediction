```python
import pandas as pd
from sklearn.model_selection import train_test_split #import hàm đề chia dataset
from sklearn.linear_model import LogisticRegression #import thuật toán logistic regression
from sklearn.metrics import accuracy_score, classification_report #import hàm kiểm tra độ chính xác
import joblib
from sklearn.preprocessing import StandardScaler

# Đưa dataset vào
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")

X = df.drop("DEATH_EVENT", axis = 1) # X là những cột dùng để kiểm tra (tất cả bảng ngoại trừ cột death_event)
y = df["DEATH_EVENT"] # y là cột death_event, là kết quả cần kiểm tra


# chia X, y ra theo tỷ lệ 20 / 80, 80 để kiểm tra, 20 để làm kết quả đối chiếu
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3
)
# chuẩn hóa dữ liệu (khoảng cách số học của dữ liệu lớn làm cho logistic regression dễ sai sót)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter = 1000) # Khai báo model

model.fit(X_train_scaled, y_train) # đưa data vào để train

y_pred = model.predict(X_test_scaled) # từ model vừa train để dự đoán death event

accuracy = accuracy_score(y_test, y_pred) # đối chiếu kết quả dự đoán với kết quả thực
print('Độ chính xác:', accuracy)

report = classification_report(y_test, y_pred)
print('Báo cáo phân loại:\n', report)

joblib.dump(model, 'logistic_model.pkl') # lưu model lại với tên logistic_model.pkl
joblib.dump(scaler, 'scaler.pkl') # lưu hàm chuẩn hóa đầu vào với tên scaler.pkl

print('Lưu thành công!')



```
